```python
import pandas as pd
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
bike_data_df = pd.read_csv('JC-201908-citibike-tripdata.csv')
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.
### Note that this will create a new column in the dataframe, rather than changing the 
### values in the original Series.
bike_data_df['datetime_column'] = pd.to_datetime(bike_data_df['tripduration'], unit='s')
bike_data_df['datetime_column'] = bike_data_df['datetime_column'].dt.strftime('%H:%M:%S')
### Note that this technically returns a string value, not a datetime value.
```

```python
# 4. Check the datatypes of your columns. 
type(bike_data_df['datetime_column'][0])
```

```python
# 5. Export the Dataframe as a new CSV file without the index.
bike_data_df.to_csv('bike_data_df.csv', index=False)
```
